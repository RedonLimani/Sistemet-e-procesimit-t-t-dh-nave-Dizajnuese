```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Initialize Selenium WebDriver (requires browser driver installed)
driver = webdriver.Chrome()  # Or another browser driver
driver.get('https://www.kaggle.com/datasets/mitishaagarwal/patient')

# Give the page time to load
time.sleep(5)

# JavaScript code to scroll to a specific element
js_code = "arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});"

# Locating the WebElement we want to scroll to
element = driver.find_element(By.CLASS_NAME, "nTRvB")

# Execute the JavaScript code to scroll to the element
driver.execute_script(js_code, element)
time.sleep(1)

# 3 clicks to display all columns
arrow = driver.find_element(By.CLASS_NAME, "nTRvB")
arrow.click()
time.sleep(1)

checkbox = driver.find_element(By.CLASS_NAME, "PrivateSwitchBase-input")
checkbox.click()
time.sleep(1)

applyBtn = driver.find_element(By.CLASS_NAME, "euFqtC")
applyBtn.click()
time.sleep(1)

# Get the rendered page source
page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

# list to hold descriptors for each column, list of columns then both merged on dict
descriptors = []
columns = []
colDesc = dict()

# Scrape data, parsing the data, removing html tags and newlines "\n"
for data in soup.find_all('p', class_="sc-PGZt"):  
    formated_data = data.get_text().replace("\n", "")
    descriptors.append(formated_data)
for data in soup.find_all('span', class_="sc-ebhTVr"):  
    formated_span_data = data.get_text()
    columns.append(formated_span_data)
    
# Pair up columns and their descriptions
for i in range(min(len(columns),len(descriptors))):
    colDesc[columns[i]] = descriptors[i]
    
# Print data just to see
print(len(colDesc))
print(colDesc)

# Close the driver
driver.quit()
```

```python
from sklearn.feature_selection import VarianceThreshold

def remove_low_variance_columns(df, threshold=0.01):
    """
    Removes columns with variance below a given threshold.
    
    Parameters:
        df (pd.DataFrame): Input DataFrame.
        threshold (float): Variance threshold (default: 0.01).

    Returns:
        pd.DataFrame: DataFrame with low-variance columns removed.
    """
    selector = VarianceThreshold(threshold)
    numeric_df = df.select_dtypes(include=['float', 'int'])
    
    selector.fit(numeric_df)
    selected_columns = numeric_df.columns[selector.get_support()]
    return df[selected_columns]

df_reduced = remove_low_variance_columns(df)
#df_reduced.info()
for i in df.columns:
    if i not in df_reduced.columns:
        print(i)

```

```python
#label encoding
from scipy.stats import chi2_contingency
import numpy as np

def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2 = chi2_contingency(contingency_table)[0]
    n = contingency_table.sum().sum()
    phi2 = chi2 / n
    r, k = contingency_table.shape
    return np.sqrt(phi2 / min((r - 1), (k - 1)))

# Example
print(cramers_v(df['apache_3j_bodysystem'], df['apache_2_bodysystem']))
#heqim apache 2 bodysystem pershkak te ngjashmeris me apache 3 bodysystem duke larguar redudanc

df=df.drop('apache_2_bodysystem',axis='columns')
df.shape
```

```python
def removeOutlier(df,clm):
    
# Calculate the upper and lower limits
    Q1 = df[clm].quantile(0.25)
    Q3 = df[clm].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 10*IQR
    upper = Q3 + 10*IQR
    
    # Create arrays of Boolean values indicating the outlier rows
    filtered_df = df[(df[clm] > lower) & (df[clm] < upper)]
    
    
    # Print the new shape of the DataFrame
    print("New Shape: ", filtered_df.shape)
    return filtered_df

for i in floatclm2:
    df=removeOutlier(df,i)
    
    
```

```python
def folding(df,n):
    folds = []
    dfcopy = df.copy()
    foldsize = dfcopy.shape[0] // n
    indices = np.random.permutation(dfcopy.index)
    
    for i in range(n):
        fold_indices = indices[i * foldsize: (i + 1) * foldsize]
        fold = dfcopy.iloc[fold_indices]
        folds.append(fold)
    
    return folds
```

```python
def accuarcy_func(desired,output):
    correct=0
    for i in range(len(desired)):
        if desired.iloc[i] == output[i]:
            correct += 1
    return correct / float(len(desired)) * 100.0
```

```python
def evaluate(df, algorithm, n, *args):
    folds = folding(df, n)
    scores = list()
    for fold in folds:
        traindf = [f for f in folds if not f.equals(fold)]
        train_set = pd.concat(traindf)
        test_set = fold.copy().drop(["hospital_death"], axis=1)
        
        predicted = algorithm(train_set, test_set, *args)
        desired = fold["hospital_death"]
        correctness = accuarcy_func(desired, predicted)
        scores.append(correctness)
    return scores
```

```python
def knn(train,test,n):
    predictions=[]
    for i in range(test.shape[0]):
        predictions.append(kojshit(train,test.iloc[i],n))
    return predictions
```
