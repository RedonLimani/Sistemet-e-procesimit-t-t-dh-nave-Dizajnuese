```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("dataset.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Initialize Selenium WebDriver (requires browser driver installed)
driver = webdriver.Chrome()  # Or another browser driver
driver.get('https://www.kaggle.com/datasets/mitishaagarwal/patient')

# Give the page time to load
time.sleep(5)

# JavaScript code to scroll to a specific element
js_code = "arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});"

# Locating the WebElement we want to scroll to
element = driver.find_element(By.CLASS_NAME, "nTRvB")

# Execute the JavaScript code to scroll to the element
driver.execute_script(js_code, element)
time.sleep(1)

# 3 clicks to display all columns
arrow = driver.find_element(By.CLASS_NAME, "nTRvB")
arrow.click()
time.sleep(1)

checkbox = driver.find_element(By.CLASS_NAME, "PrivateSwitchBase-input")
checkbox.click()
time.sleep(1)

applyBtn = driver.find_element(By.CLASS_NAME, "euFqtC")
applyBtn.click()
time.sleep(1)

# Get the rendered page source
page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

# list to hold descriptors for each column, list of columns then both merged on dict
descriptors = []
columns = []
colDesc = dict()

# Scrape data, parsing the data, removing html tags and newlines "\n"
for data in soup.find_all('p', class_="sc-PGZt"):  
    formated_data = data.get_text().replace("\n", "")
    descriptors.append(formated_data)
for data in soup.find_all('span', class_="sc-ebhTVr"):  
    formated_span_data = data.get_text()
    columns.append(formated_span_data)
    
# Pair up columns and their descriptions
for i in range(min(len(columns),len(descriptors))):
    colDesc[columns[i]] = descriptors[i]
    
# Print data just to see
print(len(colDesc))
print(colDesc)

# Close the driver
driver.quit()
```

```python
### largimi i kolonova qe nuk na sherbjen pershkak qe jan te gjitha unike 
idcolumns= ['encounter_id', 'patient_id', 'hospital_id', 'icu_id']

df=df.drop(idcolumns,axis=1)

# Qita e kom shtu un(libi), i njejti funksion veq tash osht per Dictionary
for i in idcolumns:
    colDesc.pop(i)

# print just to check
print(colDesc)
```

```python
## hjeqim rreshtat e njejt
df.drop_duplicates(inplace = True) 
print(df.duplicated())
```

```python
## i shfaqim kolonat qe jane te tipit float
floatclm=[]
for i in df.columns:
    if isinstance(df[i][0],float):
        floatclm.append(i)

print(floatclm)
```

```python
## Tek kolonat ne vijim i zevendsojm vlerat qe mungojn me mesateren e atyre vlerave
print(df['age'][28])
mesatareclm=['age','bmi','height','weight']
for i in mesatareclm:
    df[i]=df[i].fillna(round(df[i].mean(),0))

print(df['age'][28])
```

```python
### largimi i kolonave qe kan numer te madh te te dhenave null
lowrows_columns=[]
bar=df['apache_4a_hospital_death_prob'].count()
for i in df.columns:
    if(df[i].count()) < bar:
        lowrows_columns.append(i)

df=df.drop(lowrows_columns, axis=1)
```

```python
## shqyrtojm korrelacionin e kolonova me njera tjetren , ato me korrelacion te lart jane redudance e panevojshme
dc=df[floatclm[12:-6]]
corr_matrix = dc.corr() 
high_corr = corr_matrix[corr_matrix > 0.95 ]
plt.figure(figsize=(100, 80))
sns.heatmap(high_corr, annot=True)
```

```python
## largojm kolonet ne viijim pershkak te korrelacionit te lart dhe numrit me te vogel te te dhenave
noninvasive=[]
for i in df.columns:
    if "noninvasive" in i :
        noninvasive.append(i)
df=df.drop(noninvasive,axis=1)
```

```python
# shfaqim kolonat qe kane objekte(string) si vlera 
objclm=[]
for i in df.columns:
    if not isinstance(df[i][0],int) and not isinstance(df[i][0],float) :
        objclm.append(i)
print(objclm)
```

```python
# per kolonat ne vijim vendosim moden e kolones ne vlerat qe mungojn
modeclm=['ethnicity','gender','icu_admit_source','icu_type']
for i in modeclm:
    df[i]=df[i].fillna(df[i].mode())

#print(df[28])
```
